###Control Statements

Control statements in Python are used to direct the flow of a program's execution based on specific conditions or repeated logic. They allow the code to make decisions, skip sections, or repeat tasks efficiently.

###If Statements (Conditional Branching)
The if statement is used to execute a block of code only if a specified condition is true. Python uses elif (short for else if) to check multiple conditions and else to catch anything that wasn't caught by the preceding conditions.



### Indentation
In Python, indentation refers to the spaces or tabs at the beginning of a code line. While in other languages indentation is purely for readability, in Python it is a strict syntactic requirement used to define the structure and scope of code blocks. For example, statements within an `if` block, `for` loop, or `while` loop must be indented at the same level. Incorrect indentation will lead to `IndentationError`.

###For Loops (Iterative Control)
A 'for' loop is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string). It executes a set of statements once for each item in the sequence, making it ideal for tasks with a known number of repetitions.


###While Loops (Conditional Iteration)
A while loop repeatedly executes a target statement as long as a given condition remains true. It is primarily used when the number of iterations is not known beforehand and depends on dynamic factors during runtime.

### Loop Control Statements
These keywords are used to change the behavior of loops mid-execution:
* break: Immediately terminates the loop and moves to the next section of code outside the loop.
* continue: Skips the current iteration and jumps to the next cycle of the loop.
* pass: A null statement used as a placeholder when a statement is syntactically required but no action is needed.


###Lists
Lists are one of the most versatile data structures in Python, used to store a collection of items in a single variable. They are ordered, changeable (mutable), and allow for duplicate values.

Creation: Defined by placing elements inside square brackets [], separated by commas.

Indexing: Items are accessed by their position, starting at 0 for the first element.

Flexibility: A single list can contain different data types, such as integers, strings, and even other lists.


### Dictionaries
A dictionary is a collection used to store data values in key:value
pairs. They are ordered (as of Python 3.7+), changeable, and do not allow duplicate keys, making them ideal for labeled data.

Structure: Defined with curly braces {}, where each key is separated from its value by a colon :.

Access: Values are retrieved by referencing their unique key rather than an index number.

Use Case: Perfect for representing real-world objects, like a user profile or a product catalog.

### Demonstrating break and continue

break is used to exit a loop prematurely, while continue is used to skip the current iteration and move to the next one.

```python
# Example using 'break'
print("Demonstrating 'break':")
for i in range(1, 10):
    if i == 5:
        print(f"Breaking loop when i is {i}")
        break  # Exit the loop entirely
    print(f"Current number (break): {i}")

print("\nDemonstrating 'continue':")
# Example using 'continue'
for i in range(1, 10):
    if i % 2 == 0:  # If 'i' is even
        print(f"Skipping even number (continue): {i}")
        continue  # Skip to the next iteration
    print(f"Current odd number (continue): {i}")
```

```python
# Python code to print a multiplication table

# Outer loop for the first multiplier (rows)
for i in range(1, 11):
    # Inner loop for the second multiplier (columns)
    for j in range(1, 11):
        # Print the product, formatted to 5 spaces, without a newline
        print("{:5}".format(i * j), end=" ")
    # Move to the next line after each row is complete
    print()
```

###Comprehensions in Python
Comprehensions provide a concise and expressive way to create new sequences (such as lists, dictionaries, and sets) from existing iterables. They replace traditional for loops with a single line of code, making the script more readable and often more efficient.

List Comprehension
List comprehension is the most common form. It allows you to transform or filter a list in one step.

Basic Syntax: [expression for item in iterable]

With Logic: [expression for item in iterable if condition]

```python
# Traditional loop for squaring numbers
# Define a list of numbers
nums = [0, 1, 2, 3, 4, 5]
# Initialize an empty list to store squared numbers
sq_n = []
# Loop through each number in the 'nums' list
for i in nums:
    # Calculate the square of the number and add it to the new list
    sq_n.append(i ** 2)

# Print the list of squared numbers
print(sq_n)
```

```python
# List Comprehension for squaring numbers
# Use a list comprehension to square each number and create a new list
print([x**2 for x in  nums ])
```

```python
# Filter elements using a traditional loop
# Define a list of numerical values
Cal = [80, 70, 88, 77, 66, 99, 55, 78, 77]

# Iterate through each mark in the 'Cal' list
for marks in Cal:
    # Check if the current mark is greater than 70
    if marks > 70:
        # Print marks that are above 70
        print(marks)
```

```python
# Filter elements using list comprehension
# Use a list comprehension to filter marks greater than 70
h = [mark for mark in Cal if mark > 70]
# Print the filtered list of marks
print(h)
```

###Flatten Nested List using Python


A nested list contains multiple sublists. The goal is to extract all elements from these sublists and combine them into one list.


```python
# Define the main nested list
Main_list = [[1,2,3],[4,5,6],[7,8,9]]

# Initialize an empty list to store the flattened elements
Con_List = []
# Iterate through each sublist in the Main_list
for sublist in Main_list:
    # Iterate through each element within the current sublist
    for ele in sublist:
        # Add the current element to the flattened list
        Con_List.append(ele)

# Print the flattened list
print(Con_List)
```
