### Number Guessing Game

This game generates a random number between 1 and 50, and you have a limited number of tries to guess it.

```python
import random

# Generate a random number between 1 and 100
target_number = random.randint(1, 50)

# Set the number of tries
max_tries = 7
tries_left = max_tries

print(f"I have thought of a number between 1 and 50. You have {max_tries} tries to guess it.")
```

```python
guess = None

while tries_left > 0:
    try:
        # Get user input
        guess_input = input(f"You have {tries_left} tries left. Enter your guess: ")
        guess = int(guess_input)

        # Compare the guess with the target number
        if guess == target_number:
            print(f"Congratulations! You guessed the number {target_number} correctly in {max_tries - tries_left + 1} tries!")
            break
        elif guess < target_number:
            difference = target_number - guess
            if difference > 15:
                print("Too low! A huge difference.")
            elif difference >= 5:
                print("Too low!")
            else:
                print("Nearly there, but too low!")
        else: # guess > target_number
            difference = guess - target_number
            if difference > 15:
                print("Too high! A huge difference.")
            elif difference >= 5:
                print("Too high!")
            else:
                print("Nearly there, but too high!")

        tries_left -= 1

    except ValueError:
        print("Invalid input. Please enter an integer.")

    # If no tries left and still incorrect
    if tries_left == 0 and guess != target_number:
        print(f"Sorry, you ran out of tries. The number was {target_number}.")
```
