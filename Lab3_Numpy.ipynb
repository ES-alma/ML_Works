### NumPy
NumPy (short for Numerical Python) is the fundamental open-source library for scientific computing in Python. It provides high-performance multidimensional array objects and tools for working with these arrays. While Python lists are flexible, NumPy arrays are specifically designed for speed and memory efficiency when handling large datasets.

### The N-Dimensional Array (ndarray)
The core feature of NumPy is the ndarray. Unlike a standard Python list, a NumPy array contains elements of the same data type, allowing the computer to process the data much faster.

1D Array:  A single row of numbers (Vector).

2D Array:  A table of rows and columns (Matrix).

3D Array:  A collection of matrices (Tensor).



### Key Features
*  Vectorization:  Allows you to perform mathematical operations on entire arrays at once without writing for loops. For example, array * 2 multiplies every element by 2 instantly.
*  Broadcasting:  A powerful mechanism that allows NumPy to work with arrays of different shapes during arithmetic operations.
*  Mathematical Functions:  Includes a massive library of built-in functions for linear algebra, statistics, Fourier transforms, and random number generation.

### Comparison: Python Lists vs. NumPy Arrays
| Feature | Python List | NumPy Array |
| :--- | :--- | :--- |
|  Data Types  | Can mix types (int, str, etc.) | Homogeneous (all same type) |
|  Speed  | Slower for large data | Significantly faster (C-based) |
|  Memory  | High consumption | Highly memory efficient |
|  Operations  | Limited math support | Advanced mathematical


### Common NumPy Operations
*  Creation:  `np.array([1, 2, 3])`, `np.zeros((3,3))`, or `np.arange(0, 10)`.
*  Reshaping:  Changing the structure of data (e.g., turning a 1D list of 12 items into a 3x4 matrix) using `.reshape()`.
*  Slicing:  Accessing specific parts of an array using `array[start:stop]`.
*  Aggregates:  Quickly finding the `np.mean()`, `np.median()`, `np.sum()`, or `np.std()` of a dataset.



### Why use NumPy?
NumPy is the "engine" that powers almost every other Python data library.  Pandas  is built on top of NumPy,  Scikit-learn  uses NumPy arrays for machine learning inputs, and  Matplotlib  uses them for plotting. Without NumPy, data science in Python would be incredibly slow.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array 'm' from 0 to 20 with a step of 2.
m = np.arange(0, 20, 2)
# Calculate a linear function of 'm' (y = 2x + 5).
l_m = 2*m + 5

# Create a new figure for the plot.
plt.figure()
# Set the super title for the figure.
plt.suptitle("Linear Function Plot")

# Create a subplot within a 2x2 grid, selecting the first position.
plt.subplot(2, 2, 1)
# Plot the linear function 'l_m' against 'm'.
plt.plot(m, l_m)
# Set the title for this subplot.
plt.title("y = 2x + 5")

# Display the plot.
plt.show()
```

```python
### Matplotlib.pyplot

# Import the Matplotlib plotting library.
import matplotlib.pyplot as plt

# Create an array 'x' for the range 0 to 5*pi with a step of 0.1.
x = np.arange(0,5*np.pi,0.1)
# Calculate sine and cosine values for 'x'.
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the sine function.
plt.plot(x,y_sin)
# Plot the cosine function.
plt.plot(x,y_cos)
# Label the x-axis.
plt.xlabel("x")
# Label the y-axis.
plt.ylabel("Functions")
# Add a title to the plot.
plt.title("Sin and Cos Graph")
# Display the plot.
plt.show()
```

```python
# Create a linearly spaced array 'x' from 0 to 4*pi with 200 points.
x = np.linspace(0, 4*np.pi, 20)
# Calculate sine of x squared and cosine of x divided by 2.
y_sin = np.sin(x**2)
y_cos = np.cos(x/2)

# Create a 2x1 grid of subplots, and select the first subplot.
plt.subplot(2,1,1)

# Create a scatter plot for the sine of x squared function.
plt.scatter(x, y_sin, color = "darkblue")
# Label the x-axis for the first subplot.
plt.xlabel("x")
# Label the y-axis for the first subplot.
plt.ylabel("Sin(x^2) Function")
# Add a title to the first subplot.
plt.title("Trig_Sin(x^2) Function")

# Select the second subplot.
plt.subplot(2,1,2)
# Create a line plot for the cosine of x divided by 2 function.
plt.plot(x, y_cos, color = "darkgreen")
# Label the x-axis for the second subplot.
plt.xlabel("x")
# Label the y-axis for the second subplot.
plt.ylabel("Cos(x/2) Function")
# Add a title to the second subplot.
plt.title("Trig_Cos(x/2) Function")

# Display all generated plots.
plt.show()
```

```python
# Create variable using arange: x from 1 to 9.9 with step 0.1
x = np.arange(1, 10, 0.1)

# Define different mathematical functions
y_linear = 5*x + 2      # Linear function
y_power = x**3          # Power function
y_exp = np.exp(x)       # Exponential function
y_log = np.log(x)       # Logarithmic function

# Create a figure with a 2x2 grid of subplots
plt.figure(figsize=(12,8))

# Subplot 1: Linear Function
plt.subplot(2,2,1) # Position of the plot
plt.plot(x, y_linear, color='blue')
plt.title("Linear Function")
plt.grid() # Add a grid for readability

# Subplot 2: Power Function
plt.subplot(2,2,2)
plt.plot(x, y_power, color='green')
plt.title("Power Function")
plt.grid() # Add a grid for readability

# Subplot 3: Exponential Function
plt.subplot(2,2,3)
plt.plot(x, y_exp, color='yellow')
plt.title("Exponential Function")
plt.grid() # Add a grid for readability

# Subplot 4: Logarithmic Function
plt.subplot(2,2,4)
plt.plot(x, y_log, color='red')
plt.title("Logarithmic Function")
plt.grid() # Add a grid for readability

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show() # Display the plots
```
