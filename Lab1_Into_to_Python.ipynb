###About Python:

###Variables
In Python, variables are symbolic names that refer to values stored in memory, allowing you to store and manipulate data easily. You create a variable by assigning a value using the equals sign (e.g., variable_name = value).

###Assignment
In Python, assignment is the process of storing a value in a variable using the assignment operator "=", which associates the value on the right with the variable name on the left. For example, the statement x = 5 assigns the value 5 to the variable x

###Operators
Operators in Python are special symbols that perform operations on variables and values, allowing for calculations, comparisons, and control of program flow. There are several types of operators, including arithmetic, assignment, comparison, logical, and bitwise operators.

###print() function
In Python, the print() function is used to display output to the console. You can pass one or more items to it, and it will convert them to strings and print them, separating multiple items with a space by default.

##Data Types
Python automatically categorizes data into types based on the value assigned. Common types include:

Integers (int): Whole numbers (e.g., 10, -3).

Floats (float): Decimal numbers (e.g., 10.5, 3.14).

Strings (str): Text wrapped in quotes (e.g., "Hello").

Booleans (bool): Logical values, either True or False.

### Indexing and Slicing in Python

**What is Indexing and Slicing?**

Indexing allows you to access individual elements in a sequence (like strings, lists, tuples) using their position. Python uses zero-based indexing, meaning the first element is at index 0.

Slicing allows you to extract a portion of a sequence by specifying a start and end index. The syntax is: `sequence[start:end:step]`

- **Positive indexing**: Starts from 0 (beginning)
- **Negative indexing**: Starts from -1 (end)
- **Slicing**: `[start:end]` extracts elements from start up to (but not including) end

###Example Problem

```python
# Problem 1: Confusion Matrix Metrics
# flood prediction model - binary classification

# given values from the confusion matrix
TP = 50  # True Positives
FP = 8   # False Positives
TN = 32  # True Negatives
FN = 10  # False Negatives

# Task 1: total predictions
total_predictions = TP + FP + TN + FN
print("Total Predictions:", total_predictions)

# Task 2: calculating metrics
# accuracy formula
accuracy = (TP + TN) / (TP + FP + TN + FN)
print("Accuracy:", accuracy)

# Precision - how many predicted floods were actual floods
precision = TP / (TP + FP)
print("Precision:", precision)

# Recall (sensitivity) - how many actual floods did we catch
recall = TP / (TP + FN)
print("Recall:", recall)

# F1 score - harmonic mean of precision and recall
F1 = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", F1)

# Interpretation:
# Accuracy is 0.82 (82%) - model is correct 82% of the time
# Precision is high (0.86) - when we predict flood, we're usually right
# Recall is good (0.83) - we catch most of the actual floods
# F1 Score (0.85) - good balance between precision and recall
```

```python
# Problem 2: R² (Coefficient of Determination)
# regression model - predicting river discharge levels

# Given values
SSR = 420  # Sum of Squares Regression
SSE = 180  # Sum of Squares Error

# Task 1: calculate SST (Total Sum of Squares)
SST = SSR + SSE
print("Total Sum of Squares (SST):", SST)

# Task 2: Calculate R-squared
# R² tells us how well the model explains the data
R_squared = SSR / SST
print("R² (R-squared):", R_squared)

# Task 3: check if model explains more than 70% variance
if R_squared > 0.70:
    print("Interpretation: The model explains more than 70% of the variance.")
else:
    print("Interpretation: The model explains less than or equal to 70% of the variance.")
```

```python
# Problem 3: Mean Squared Error (MSE)
# Temperature prediction model

# Task 1: Store the squared errors
squared_errors = [2, 4, 1, 9, 4]  # squared errors for 5 observations
print("Squared errors:", squared_errors)

# Task 2: Calculate MSE
# MSE = average of squared errors
n = 5  # number of observations
sum_squared_errors = 2 + 4 + 1 + 9 + 4  # adding all squared errors
MSE = sum_squared_errors / n

# Interpretation:
# MSE of 4.0 means the average squared error is 4
# Lower MSE = better model predictions
# This tells us on average how far off our temperature predictions are

# Task 3: print result
print("Mean Squared Error (MSE):", MSE)
```

```python
# Problem 4: Cook's Distance - Influential Observation
# checking if an observation has too much influence on regression

# Given values
e_squared = 9    # Squared residual
MSE = 3          # Mean Squared Error
p = 3            # Number of predictors
h = 0.2          # Leverage value
n = 25           # Sample size

# Task 1: Compute Cook's Distance
# using the formula D_i = (e_i^2 / (p * MSE)) * (h_ii / (1 - h_ii)^2)
D_i = (e_squared / (p * MSE)) * (h / ((1 - h) ** 2))
print("Cook's Distance (D_i):", D_i)

# Task 2: Compare with threshold 4/n
threshold = 4 / n
print("Threshold (4/n):", threshold)

# Task 3: check if observation is influential
if D_i > threshold:
    print("Decision: The observation is INFLUENTIAL (D_i > 4/n)")
else:
    print("Decision: The observation is NOT INFLUENTIAL (D_i <= 4/n)")
```
